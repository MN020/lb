# Data Augmentation

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
# Define augmentation settings
datagen = ImageDataGenerator(
     rotation_range=30,           # Random rotation between 0–30°
    width_shift_range=0.1,       # Shift along width (10%)
    height_shift_range=0.1,      # Shift along height (10%)
    zoom_range=0.2,              # Zoom in/out by 20%
    horizontal_flip=True,        # Randomly flip images horizontally
    vertical_flip=False,         # You can enable this for certain datasets
    shear_range=0.2,             # Shearing transformations
    brightness_range=[0.7, 1.3], # Vary image brightness
    fill_mode='nearest'          # Filling strategy for pixels after rotation/shift
)
```

```python
# Load your image (replace 'your_image.jpg' with the actual path)
img = load_img('butterfly.png')  # load image
x = img_to_array(img)             # convert to numpy array
x = np.expand_dims(x, axis=0)     # add batch dimension
```

```python
# Generate and plot augmented images
plt.figure(figsize=(10, 10))
for i in range(9):
    batch = datagen.flow(x, batch_size=1)
    image = next(batch)[0].astype('uint8')
    plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.axis('off')
plt.tight_layout()
plt.show()
```
