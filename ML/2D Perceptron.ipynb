# 2D Perceptron

```python
import numpy as np
import matplotlib.pyplot as plt

# ML Lab - Perceptron
# Sample Code by Prof Dr Aditya Abhyankar
# Simulation of Perceptron Learning

# Input data
x = np.array([[1, 1, 0, 0], [1, 0, 1, 0]])  # Input vectors (2x4)
t = np.array([1, -1, -1, -1])  # Target labels
w = np.array([0, 0])  # Initial weights
b = 0  # Initial bias
alpha = 1  # Learning rate
theta = 0.2  # Threshold
yes = 1  # Flag for training loop
epoch = 0  # Initialize epoch counter

while yes:
    yes = 0
    for i in range(4):
        plt.clf()  # Clear the current figure before each plot
        # Plot the input data points
        plt.plot(x[0, 0], x[1, 0], 'or', markersize=20, markerfacecolor=[0, 0, 1])  # Blue
        plt.plot(x[0, 1], x[1, 1], 'or', markersize=20, markerfacecolor=[1, 0, 0])  # Red
        plt.plot(x[0, 2], x[1, 2], 'or', markersize=20, markerfacecolor=[1, 0, 0])  # Red
        plt.plot(x[0, 3], x[1, 3], 'or', markersize=20, markerfacecolor=[1, 0, 0])  # Red

        # Plot the decision boundary (a line)
        m = -(w[0] / w[1]) if w[1] != 0 else 0
        c = -b / w[1] if w[1] != 0 else 0
        x1 = np.linspace(-2, 2, 100)
        if theta == 0:
            x2 = m * x1 + c
            plt.plot(x2, x1, 'b')
        else:
            x2 = m * x1 + c + theta
            x3 = m * x1 + c - theta
            plt.plot(x2, x1, 'r')
            plt.plot(x3, x1, 'r')

        plt.axis([-2, 2, -2, 2])
        plt.pause(0.1)

        # Perceptron update rule
        yin = b + x[0, i] * w[0] + x[1, i] * w[1]
        if yin > theta:
            y = 1
        elif yin >= -theta:
            y = 0
        else:
            y = -1
        
        if y != t[i]:
            yes = 1  # If output doesn't match target, continue training
            for j in range(2):
                w[j] += t[i] * x[j, i] * alpha  # Update weights
            b += t[i] * alpha  # Update bias

    epoch += 1  # Increment epoch counter

print('Perceptron')
print('Final Weight Matrix:')
print(w)
print('Final bias Values:')
print(b)
print('Total epochs used:')
print(epoch)

# Final plot
plt.figure(2)
plt.plot(x[0, 0], x[1, 0], 'or', markersize=20, markerfacecolor=[0, 0, 1])  # Blue
plt.plot(x[0, 1], x[1, 1], 'or', markersize=20, markerfacecolor=[1, 0, 0])  # Red
plt.plot(x[0, 2], x[1, 2], 'or', markersize=20, markerfacecolor=[1, 0, 0])  # Red
plt.plot(x[0, 3], x[1, 3], 'or', markersize=20, markerfacecolor=[1, 0, 0])  # Red

# Plot the decision boundary (a line)
m = -(w[0] / w[1]) if w[1] != 0 else 0
c = -b / w[1] if w[1] != 0 else 0
x1 = np.linspace(-2, 2, 100)
if theta == 0:
    x2 = m * x1 + c
    plt.plot(x2, x1, 'r')
else:
    x2 = m * x1 + c + theta
    x3 = m * x1 + c - theta
    plt.plot(x2, x1, 'r')
    plt.plot(x3, x1, 'r')

plt.axis([-2, 2, -2, 2])
plt.show()
```
